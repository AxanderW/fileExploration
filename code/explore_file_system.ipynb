## User will have the option to:
#### 1) copy a source directory to a target directory
#### 2) move source directory to target directory
#### 3) find all instance of a particular file 
#### 4) find all instances of a particular file and consolidate all instance of that file into one. *Note files must all have same fields

```python
# import dependencies

import shutil
import os
import pandas as pd
import numpy as np
import csv
import glob
from shutil import copyfile
from sys import exit
```

```python
combined_excel = []
RootDirTest = ""
TargetFolderTest = ""
fileName = ""

def get_src_folder():
    RootDirTest = str(input("Enter Root Directory with full path name: "))
    return RootDirTest
    # future improvement: write code to confirm selection

def get_target_folder():
    TargetFolderTest = str(input("\nEnter Target Folder with full path name: "))
    return TargetFolderTest
    
    # future improvement: write code to confirm selection

def get_filename():
    fileName = str(input("\nEnter Filename: "))
    return fileName
    
    # future improvement: write code to confirm selection

def copy_directory(rootdir,targetfolder):
    #get_src_folder()
    #get_target_folder()
    print("\nCopying directory: {} in progress...".format(rootdir,targetfolder))
    shutil.copytree(rootdir,targetfolder)
    print("\nDirectory successfully copied from {} to Target location: {}".format(rootdir,targetfolder))


def move_directory(rootdir,targetfolder):
    print("\nMoving rootdirectory: {} in progress.." .format(rootdir))
    shutil.move(rootdir,targetfolder)
    print("\nDirectory successfully moved from {} to Target location: {}".format(rootdir,targetfolder))

def find_file(rootdir,filename):
    print("\nSearching for files with name: {} in rootdirectory: {}".format(filename,rootdir))
    print("\nProcessing in Progress...\n")
    print("\nFiles with name: {} found in locations below: \n".format(filename))
    for root,dirs, files in os.walk((os.path.normpath(rootdir)), topdown=False):
        for name in files:
            if name.startswith(filename):
                print("-----------------")
                print("Found")
                SourceFolder = os.path.join(root,name)
                #shutil.copy2(SourceFolder,TargetFolder) # comment this line out to perform another function
                print(SourceFolder)
    

def find_consolid_file(rootdir,filename,targetfolder):
    print("Searching for files with name: {} in rootdirectory: {} \n\nConsolidated file will output to location: {}\n".format(filename,rootdir,targetfolder))
    print("Processing in Progress...\n")
    for root,dirs, files in os.walk((os.path.normpath(rootdir)), topdown=False):
        for name in files:
            if name.startswith(filename):
                print("-----------------")
                print("Found")
                SourceFolder = os.path.join(root,name)
                #shutil.copy2(SourceFolder,TargetFolder) # comment this line out to perform another function
                print(SourceFolder)
                combined_excel.append(SourceFolder)
                
    print("\nConsolidating files with name: {} in progress...".format(filename))
    # consolidate all files in the list
    consolidated_files = pd.concat(pd.read_excel(f, sort=False) for f in combined_excel)
    print("\nConsolidating complete")
    
    #Export files to target
    print("\nExporting file...")
    # export to excel
    consolidated_files.to_excel("{}/Consolidated_File.xlsx".format(targetfolder),
                         sheet_name='CONSOLIDATED_FILE', na_rep='',float_format="%.0f",
                  header=True, index=False)
    print("\nExport comeplete")
    print("\nFile Preview: \n\n")
    print(consolidated_files.head())
```

```python
print("Please select an option below: \n\n1. Copy a source directory to a target location."+
     "\n2. Move source directory to a target location." +
     "\n3. Find all instances of a file in a rootdirectory." +
     "\n4. Find all instance of a file in a rootdirectory, consolidate the files and output files to a target location.")
selection = str(input("\n\nSelection: ")).strip()

if (selection == "1" or selection == "1." or selection == "#1"):
    #get_src_folder()
    #get_target_folder()
    RootDirTest = str(input("Enter Root Directory with full path name: "))
    TargetFolderTest = str(input("\nEnter Target Folder with full path name: "))
    copy_directory(RootDirTest,TargetFolderTest)

elif (selection == "2" or selection == "2." or selection == "#2"):
    #get_src_folder()
    #get_target_folder()
    RootDirTest = str(input("Enter Root Directory with full path name: "))
    TargetFolderTest = str(input("\nEnter Target Folder with full path name: "))
    move_directory(RootDirTest,TargetFolderTest)

elif (selection == "3" or selection == "3." or selection == "#3"):
   # get_src_folder()
    #get_filename()
    RootDirTest = str(input("Enter Root Directory with full path name: "))
    fileName = str(input("\nEnter Filename: "))
    find_file(RootDirTest,fileName)
    
elif (selection == "4" or selection == "4." or selection == "#4"):
    #get_src_folder()
    #get_target_folder()
    #get_filename()
    RootDirTest = str(input("Enter Root Directory with full path name: "))
    TargetFolderTest = str(input("\nEnter Target Folder with full path name: "))
    fileName = str(input("\nEnter Filename: "))
    find_consolid_file(RootDirTest,fileName,TargetFolderTest)
else:
    print("Please make a selection")
    selection = str(input("\n\nSelection: ")).strip()
```
